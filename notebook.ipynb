```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")

df = pd.read_csv("students.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['G3'], bins=20, kde=True)

plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")

plt.show()
```

```python
df.corr(numeric_only=True)['G3'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['studytime'], y=df['G3'])

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['failures'], y=df['G3'])

plt.title("Failures vs Final Grade")
plt.xlabel("Number of Past Failures")
plt.ylabel("Final Grade")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['sex'], y=df['G3'])

plt.title("Gender vs Final Grade")
plt.xlabel("Gender")
plt.ylabel("Final Grade")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['internet'], y=df['G3'])

plt.title("Internet Access vs Final Grade")
plt.xlabel("Internet Access")
plt.ylabel("Final Grade")

plt.show()
```
